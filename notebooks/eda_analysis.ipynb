```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../data/raw/global_cars_enhanced.csv")
```

```python
# Bar chart of Efficiency_Score by Body_Type
plt.figure(figsize=(10, 6))
sns.barplot(x='Fuel_Type', y='Efficiency_Score', data=data)
plt.title('Average Efficiency Score by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Average Efficiency Score')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Price_Category by Efficiency_Score
plt.figure(figsize=(10, 6))
sns.boxplot(x='Price_Category', y='Efficiency_Score', data=data)
plt.title('Efficiency Score by Price Category')
plt.xlabel('Price Category')
plt.ylabel('Efficiency Score')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
